```python
import networkx as nx
import osmnx as ox

%matplotlib inline
ox.config(log_console=True)
ox.__version__
```

```python

# download/model a street network for some city then visualize it
G = ox.graph_from_place("Piedmont, California, USA", network_type="drive")
fig, ax = ox.plot_graph(G)
```

```python

# convert your MultiDiGraph to an undirected MultiGraph
M = ox.utils_graph.get_undirected(G)

# convert your MultiDiGraph to a DiGraph without parallel edges
D = ox.utils_graph.get_digraph(G)
```

```python
fig, ax = ox.plot_graph(G)
```

地名で日吉の道路を検索

```python
G = ox.graph_from_place("Hiyoshi, Kanagawa, Japan", network_type="drive")
fig, ax = ox.plot_graph(G)
```

```python
#緯度経度の範囲を指定し自動車が通行できる道路をダウンロード
G = ox.graph_from_bbox(35.553685, 35.563685, 139.644954, 139.664954, network_type='drive')
G_projected = ox.project_graph(G)
ox.plot_graph(G_projected)
```

```python
#日吉駅の緯度経度から750mの範囲の全てのタイプの道路ネットワーク取得
G = ox.graph_from_point((35.553685, 139.644954), dist=750, network_type='all')
ox.plot_graph(G)
```

```python
#日吉駅の緯度経度から250mの範囲の全てのタイプの道路ネットワーク取得
G = ox.graph_from_point((35.553685, 139.644954), dist=250, network_type='all')
ox.plot_graph(G)
```

Gはnetworkxパッケージ(https://networkx.org/)でのネットワークを表す。
